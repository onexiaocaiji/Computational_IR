# The H<sub>2</sub>O molecule: Vibrational Frequencies and IR Spectrum

The goal of performing an analysis of vibrational frequencies is twofold:

1) to confirm that the optimised geometry is indeed a minimum;

2) to calculate the vibrational spectrum.


It is important to highlight the following:
**a frequencies calculation requires that a geometry corresponds to a stationary point in the potential energy surface.**

Therefore, the starting point for a frequencies calculation is to open the file .chk obtained at the end of geometry optimisation.

#### Peak intensity
As discussed in the year 1 Introduction to Spectroscopy module, the relative intensity of different peaks will depend on the magnitude of the transition dipole moment $\mu_{fi}$. In the case of a diatomic molecule we saw that for small displacements we could write:

$$\mu_{fi}=\left.\frac{d \mu(x)}{dx}\right|_{x=0} \int_{-\infty}^{\infty}\chi_f(x)\ x \ \chi_i(x)dx$$

where $\mu(x)$ is how the magnitude of the dipole moment changes with the displacement $x$ of the equilibrium position $(x=0)$, and $\chi_f(x)$ and $\chi_i(x)$ are the final and initial vibration functions. The derivative in front of the integral shows that the absorption intensity is higher if the displacement induces a bigger change in the dipole moment magnitude.

Although the above equation was derived for diatomics, it also applies for individual vibrational modes in polyatomic molecules, with $x$ being a displacement of a specific vibrational modes. Vibrational modes resulting in a bigger change in dipole moment magnitude will result in more intense peaks. Conversely, vibrational modes that preserve the symmetry of the molecule and do not result in a change in dipole moment will not have observed transitions in the infrared spectrum.

In the "Vibrations" menu, you can click on "Show Dipole Derivative Unit Vector", which indicates how the dipole of the molecule changes as the molecule vibrates according to each vibrational mode. Note that the represented vector shows only the direction of the change and does not represent the magnitude of the change (unit vector). To plot the dipole moment itself (not the change), you can go to "Results" > "Charge Distribution ..." > "Dipole Moment".

#### Peak width
In the computed spectrum, each peak is represented by a gaussian function with a broadening of 30&nbsp;cm<sup>-1</sup> (see "IR Peak at Half-Width at Half Height" in the Properties window of the Plot). 

What we have calculated is the position of the peaks, the broadening is artificial and added for helping the comparison with experimental spectra, which can have different broadening.

If you click in "Plots", you can save the IR spectrum as a figure and, if necessary, you can change the properties.

### Comparison to experimental spectra
Note that the computed spectrum is for an isolated molecule; this spectrum is best compared with that obtained in the gas phase. But even in the gas phase, one may often observe rotational structure on the experimental spectrum, which is not taken into account in the calculation.

Care must be taken when comparing with liquid or solid samples, where intermolecular interactions are significant and cause a shift in the peak frequencies, intensity and broadening.
