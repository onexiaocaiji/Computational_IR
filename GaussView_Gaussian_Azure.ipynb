# How to start GaussView and Gaussian using Microsoft Azure Lab

This way of running is using a virtual machine on Microsoft cloud.
It is the suggested access method, if you are based in Asia and 
if your remote desktop connection to Imperial is too slow.

To start a remote connection to a Microsoft Azure Lab, follow the instructions in the invitation email. 

Once you click on the link in the email,
you will be directed to a webpage with a box corresponding to the virtual machine shared with you.

In the bottom left, slide the toggle from left to right to start the virtual machine (VM);
it will take around 5 minutes.
(You can set up a timer instead of waiting)
When the VM is running, click on the icon with the shape of a computer
at the bottom right, near the 3 vertical dots,
a file with the extension rdp will be downloaded: IC_Chemistry_Asia_Pacific.rdp 

Start a remote desktop connection using the previous file.

A window will appear asking to "Accept Certificate?",
press "Yes".
A window with title "Enter authentication credentials" will appear,
change the username into "chemistry" by removing "~/",
Type Imperial2122 as passwd.

After you have launched GaussView, the following windows will pop up:

1) the Control window (with the menu bar, the tool bar and the preview window)

2) the New Molecule window (which can be also open via the menu bar under "File").
