# How to start GaussView and Gaussian

If you are on campus, you can start softwarehub as indicated below. 
Alternatively, if you are at home, to connect a PC on campus follow the instructions for a remote desktop connection (on-line / remote mode).

#### Remote Desktop Connection (on-line / remote mode)

Please follow the instruction below to start a remote desktop connection.

https://www.imperial.ac.uk/admin-services/ict/self-service/connect-communicate/remote-access/remotely-access-my-college-computer/remote-desktop-access-for-students/

Please if the remote desktop connection does not work, please get in touch with Dr G. Mallia.

## SoftwareHub

Once logged to a computer on campus, visit the link below:

https://softwarehub.imperial.ac.uk/

After you have launched GaussView from softwarehub, you will have a window entitled "Clouding Player".

Wait until the package "Gaussian 16W - GaussView 6.1" is "Ready", before clicking on the green icon "Launch".

The program status will change into "Running" as indicated below:

<img src="figures/cloudingplayes.png" width="50%">

The following will will pop up:

<img src="figures/GaussView0.png" width="50%">

Click on GaussView as highlighted above, this will open GaussView:

1) the Control window (with the menu bar, the tool bar and the preview window)

2) the New Molecule window (which can be also open via the menu bar under "File").
