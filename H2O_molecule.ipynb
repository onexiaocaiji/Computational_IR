# The H$_2$O molecule: Geometry Definition

The first step to run a simulation is to define the geometry of the molecule.

The following video describes how this is done in GaussView

GaussView 6 Tutorial 1: Building Molecules   
https://www.youtube.com/watch?v=dqKWfXsEcWs

The rest of this notebook outlines how to perform this task.
In addition, you can measure distances and bond angles in the molecule.

In the top main menu, select "Builder", the "Element Fragment" window will pop up (see bottom right window with the periodic table). 

Please pay attention also to the option, "Ring fragment" which will be needed to generate a benzene ring.

<img src="figures/GaussView1_H2Ogeom.png" width="75%">

Select the O atom and click on "New Molecule window" (bottom left window). A water molecule will be displayed.

In the bottom right window, there is also the option to select the number of bonds.

To see the atom symbol and the atom label, right-click and select "View", then "Labels" and "Symbols", as indicated below.

<img src="figures/GaussView2_H2Ogeom.png" width="75%">

If you select "Center" in the sub-menu above, the molecule will be place at the centre of the window.

The same sub-menu is available under "View" in the top main menu (same line of "File", "Edit", "View" , "Builder").

**TO UNDO**: Ctrl-Z or right-click, then Edit, then Undo.

You can use the following commands:
    
Press the left-button and move the mouse (or your finger on the touchpad) to rotate the molecule.

Press the middle-button and move the mouse, to translate the molecule up/down and left/right.

Press the right-button and move the mouse, to zoom in and zoom out.

Ctrl+z = to undo

## Structure measurements: the Inquire button

Once the geometry edition is over, click on the "Inquire" icon - the question mark (see below), this will prevent to accidentally edit and modify the current structure.

If you click on the atoms of the water molecule in "Inquire" mode, bond distances and angles will be printed.

## Symmetry

The geometry built can be symmetrised; the point symmetry of the molecule is identified. 

Go under "Edit" in the top bar menu, then select "Symmetrize".

<img src="figures/GaussView4_Symmetrize.png" width="75%">

The point group will be given in the bottom part of the "Molecule" window.
