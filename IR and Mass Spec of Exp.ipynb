```python
#IR_spectrum
import matplotlib.pyplot as plt
import numpy as np

xy_data=np.loadtxt("IR_spectrum.dat")

x,y=zip(*xy_data)

plt.figure(figsize=(10,4))
plt.plot(x,y)
plt.xlabel("$\\tilde{\\nu}/cm^{-1}$")
plt.ylabel("Absorbance")
plt.grid()
plt.savefig("IR_spectrum.png")
plt.show()
```

```python
#Mass_spectrum
import matplotlib.pyplot as plt
import numpy as np

xy_data=np.loadtxt("Mass_spectrum.dat")
x,y=zip(*xy_data)

# plot a small point  
y=np.array(y)
for i in range(len(y)):
    if y[i] < 0.5:
        y[i] = 0.5

plt.figure(figsize=(10,4))
plt.bar(x,y)

for i in range(len(x)):
    if y[i]> 2:
        plt.annotate(str(int(x[i])), xy=(x[i],y[i]), ha='center', va='bottom')

# to be annotated always
plt.annotate(str(int(x[-3])), xy=(x[-3],y[-3]), ha='center', va='bottom')
plt.annotate(str(int(x[-1])), xy=(x[-1],y[-1]), ha='center', va='bottom')

plt.xlabel("m/z")
plt.ylabel("Relative Intensity")
plt.grid()
plt.savefig("Mass_spectrum.png")
plt.show()
```
