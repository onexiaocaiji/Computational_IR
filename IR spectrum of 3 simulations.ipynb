```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normalise(x):
    return (x - np.min(x))/(np.max(x) - np.min(x))

def get_spec(file, spec_label):
    xy_data=np.loadtxt("IR_spectrum.dat")
    x,y=zip(*xy_data)

    ir_Guess = np.loadtxt(file)
    x_Guess, y_Guess, der_Guess = zip(*ir_Guess)

    plt.figure(figsize=(10,4))
    plt.plot(x,normalise(y),label='Exp')
    plt.plot(x_Guess,normalise(y_Guess),label=spec_label)
    plt.xlabel("$\\tilde{\\nu}/cm^{-1}$")
    plt.ylabel("Absorbance")
    plt.grid()
    plt.title('Comparison bewteen experimental spectrum and simulated spectrum of ' + spec_label)
    plt.legend(loc='best')
    plt.show()
    return
```

## Guess 2 spectrum

```python
get_spec('Guess_2_freq_ir.txt', 'Guess 2')
```

## Guess 5 spectrum

```python
get_spec('Guess_5_freq_ir.txt', 'Guess 5')
```

## Guess 6 spectrum

```python
get_spec('Guess_6_freq_ir.txt', 'Guess 6')
```

## Overall comparison

```python
xy_data=np.loadtxt("IR_spectrum.dat")
x,y=zip(*xy_data)

ir_Guess_2 = np.loadtxt('Guess_2_freq_ir.txt')
x_Guess_2, y_Guess_2, der_Guess_2 = zip(*ir_Guess_2)

ir_Guess_5 = np.loadtxt('Guess_5_freq_ir.txt')
x_Guess_5, y_Guess_5, der_Guess_5 = zip(*ir_Guess_5)
                        
ir_Guess_6 = np.loadtxt('Guess_6_freq_ir.txt')
x_Guess_6, y_Guess_6, der_Guess_6 = zip(*ir_Guess_6)
                    
plt.figure(figsize=(20,8))
plt.plot(x,normalise(y),label='Exp')
plt.plot(x_Guess_2,normalise(y_Guess_2)+1,label='Guess 2')
plt.plot(x_Guess_5,normalise(y_Guess_5)+2,label='Guess 5')
plt.plot(x_Guess_6,normalise(y_Guess_6)+3,label='Guess 6')
plt.xlabel("$\\tilde{\\nu}/cm^{-1}$")
plt.ylabel("Absorbance")
plt.gca().axes.yaxis.set_visible(False)
plt.grid()
plt.title('Comparison between all three spectrums with experimental spectrums')
plt.legend(loc='best')
plt.show()
```
