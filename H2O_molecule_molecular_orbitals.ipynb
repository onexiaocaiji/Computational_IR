# The H$_2$O molecule: Molecular Orbitals

In order to calculate the IR spectra, we do not need to calculate the MOs,
but we are covering this analysis here, in case you want to use them at a later stage.

Please note that this section links to the yr2 ESB comp lab and the content presented in the lectures about H$_2$O MOs by Prof Mike Bearpark.

To display the molecular orbital energies, once the file .chk is opened, go under "Tools" in the main menu and select "MOs...", this will open the window with the list of the energies for each molecular orbitals.

<img src="figures/GaussView_MO1.png" width="45%" style="float:left">
<img src="figures/GaussView_MO2.png" width="30%" style="float:middle">

To visualise the MO, go under "Results" and select "Surface Contours", the window on the left will pop up; as indicated select "New Cube", which is the format used to save the values of a molecular orbital.

<img src="figures/GaussView_MO3.png" width="60%" style="float:left">
<img src="figures/GaussView_MO4.png" width="30%" style="float:right">

Once the window on the left pops up, in the box "Orbitals:" select "Occupied", then click "OK"; the window on the left will appear with the cube information for each occupied orbital. Select "New Surface" as shown in the window on the right. The selected molecular orbital (in this case the first molecular orbital) will be visualised as a surface, more properly as an isosurface, since all the points of this surface have the same value (see value in the box below: MO = 0.02 atomic unit).

<img src="figures/GaussView_MO5.png" width="30%" style="float:left">
<img src="figures/GaussView_MO6.png" width="30%" >

Repeat this operation for all the occupied molecular orbitals, you will get an isosurface for each molecular orbital, as indicated in the left window below. The HOMO is displayed in the right window as an example (please note the red square, near the energy value).

<img src="figures/GaussView_MO7.png" width="30%" style="float:left">
<img src="figures/GaussView_MO8.png" width="30%" >
